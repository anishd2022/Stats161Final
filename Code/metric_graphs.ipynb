```python
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

numeric_cols = X_train_og.select_dtypes(include=[np.number]).columns

cols_to_check = list(numeric_cols)

print("\n--- Feature Distribution Comparison (KS Test and Histograms) ---")
for col in cols_to_check:
    print(f"\n=== Feature: {col} ===")

    # KS test: Original vs SMOTE
    if col in smoteN.columns:
        ks_sm = ks_2samp(X_train_og[col].values, smoteN[col].values)
        print(f"KS Test (Original vs SMOTE): statistic={ks_sm.statistic:.4f}, p-value={ks_sm.pvalue:.4e}")
    else:
        print(f"Column '{col}' not found in smoteN, skipping KS test for SMOTE.")

    # KS test: Original vs ADASYN
    if col in adasynN.columns:
        ks_ada = ks_2samp(X_train_og[col].values, adasynN[col].values)
        print(f"KS Test (Original vs ADASYN): statistic={ks_ada.statistic:.4f}, p-value={ks_ada.pvalue:.4e}")
    else:
        print(f"Column '{col}' not found in adasynN, skipping KS test for ADASYN.")

    plt.figure(figsize=(10, 6))
    sns.histplot(X_train_og[col], bins=50, alpha=0.5, stat='density', label='Original')
    if col in smoteN.columns:
        sns.histplot(smoteN[col], bins=50, alpha=0.3, stat='density', label='SMOTE')
    if col in adasynN.columns:
        sns.histplot(adasynN[col], bins=50, alpha=0.3, stat='density', label='ADASYN')

    plt.title(f'Distribution of {col}: Original vs SMOTE vs ADASYN')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
```

```python
def compute_dcr(real_df, syn_df, feature_cols=None, metric='euclidean', n_neighbors=1):

    if feature_cols is None:
        feature_cols = list(set(real_df.columns).intersection(set(syn_df.columns)))
        feature_cols = sorted(feature_cols)
    
    X_real = real_df[feature_cols].to_numpy()
    X_syn  = syn_df[feature_cols].to_numpy()
    

    scaler = StandardScaler()
    X_real_scaled = scaler.fit_transform(X_real)
    X_syn_scaled  = scaler.transform(X_syn)
    
    nn = NearestNeighbors(n_neighbors=n_neighbors, metric=metric)
    nn.fit(X_real_scaled)
    
    dists, _ = nn.kneighbors(X_syn_scaled)
    min_dists = dists[:, 0]

    summary = {
        'mean':   float(np.mean(min_dists)),
        'median': float(np.median(min_dists)),
        'min':    float(np.min(min_dists)),
        'max':    float(np.max(min_dists)),
        'p5':     float(np.percentile(min_dists, 5)),
        'p25':    float(np.percentile(min_dists, 25)),
        'p75':    float(np.percentile(min_dists, 75)),
        'p95':    float(np.percentile(min_dists, 95)),
    }
    
    return {
        'min_dists': min_dists,
        'summary': summary
    }

numeric_cols = X_train_og.select_dtypes(include=[np.number]).columns.tolist()

common_numeric_cols = sorted(
    set(numeric_cols).intersection(smoteN.columns).intersection(adasynN.columns)
)

print("Using", len(common_numeric_cols), "numeric columns for DCR.")

dcr_smote = compute_dcr(
    real_df=X_train_og,
    syn_df=smoteN,
    feature_cols=common_numeric_cols,
    metric='euclidean'
)

print("=== DCR: Real vs SMOTE ===")
for k, v in dcr_smote['summary'].items():
    print(f"{k:>6}: {v:.4f}")

dcr_adasyn = compute_dcr(
    real_df=X_train_og,
    syn_df=adasynN,
    feature_cols=common_numeric_cols,
    metric='euclidean'
)

print("\n=== DCR: Real vs ADASYN ===")
for k, v in dcr_adasyn['summary'].items():
    print(f"{k:>6}: {v:.4f}")

plt.figure(figsize=(8, 5))
plt.hist(dcr_smote['min_dists'], bins=50, alpha=0.6, label='SMOTE')
plt.hist(dcr_adasyn['min_dists'], bins=50, alpha=0.6, label='ADASYN')
plt.xlabel("Distance to closest real record (scaled space)")
plt.ylabel("Count")
plt.title("DCR distribution: SMOTE vs ADASYN (vs X_train_og)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
